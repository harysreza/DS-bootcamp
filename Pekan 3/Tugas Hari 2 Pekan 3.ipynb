Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

plt.plot(x, cos, linestyle = 'solid', color='red', label = 'Cosx')
plt.plot(x, cos3, linestyle = 'dashed', color='green', label = '3 Cosx')
plt.plot(x, sin, linestyle = 'dashdot', color='blue', label = 'Sin(x-90)')
plt.plot(x, sin2, linestyle = 'None', marker='x', color='yellow', label = '2 Sinx')

plt.legend(loc = 'center')

plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.title('Grafik Cos Sin')

plt.show()
```
